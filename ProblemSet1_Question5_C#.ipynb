```
#a. Write a function inside(x,y,x1,y1,x2,y2) that returns True or False
#depending on whether the point (x, y) lies in the rectangle with lower left
#corner (x1, y1) and upper right corner (x2, y2).

static bool inside(double x1, double y1, double x2, double y2, double x, double y)
{
    if (x>x1 && x<x2 && y>y1 && y<y2)
    {
        return true;
    }
        return false;     
}
Console.WriteLine("Enter the lower left corner points of first rectangle: ");
double x1 = Convert.ToDouble(Console.ReadLine());
double y1 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the upper right corner points of first rectangle: ");
double x2 = Convert.ToDouble(Console.ReadLine());
double y2 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter your points to check if its in rectangle: ");
double x = Convert.ToDouble(Console.ReadLine());
double y = Convert.ToDouble(Console.ReadLine());

Console.WriteLine(inside(x1,y1,x2,y2,x,y));

#5.b.Use function inside() from part a.to write an expression that tests whether 
#the point (1,1) lies in both of the following rectangles: one with lower left corner (0.3, 0.5) 
#and upper right corner (1.1, 0.7) and the other with lower left corner (0.5, 0.2) and upper right corner (1.1, 2)

Console.WriteLine("Enter the lower left corner points of first rectangle: ");
double x3 = Convert.ToDouble(Console.ReadLine());
double y3 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the upper right corner points of first rectangle: ");
double x4 = Convert.ToDouble(Console.ReadLine());
double y4 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the lower left corner points of second rectangle: ");
double x5 = Convert.ToDouble(Console.ReadLine());
double y5 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter the upper right corner points of second rectangle: ");
double x6 = Convert.ToDouble(Console.ReadLine());
double y6 = Convert.ToDouble(Console.ReadLine());

Console.WriteLine("Enter your points to check if its in rectangle: ");
double a = Convert.ToDouble(Console.ReadLine());
double b = Convert.ToDouble(Console.ReadLine());

bool rectangle1 = inside(x3, y3, x4, y4, a, b);
bool rectangle2 = inside(x5, y5, x6, y6, a, b);

if(rectangle1 == rectangle2)
{
    Console.WriteLine("The point lies in both the rectangles");
}
else
{
    Console.WriteLine("Sorry, The point doen't lie in both the rectangles");
}
```
